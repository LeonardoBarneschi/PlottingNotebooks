```python
from matplotlib.ticker import FormatStrFormatter
from matplotlib import ticker
from matplotlib import pyplot as plt
import matplotlib.font_manager as fm

import pandas as pd
import csv
import os
import sys
import numpy as np
import matplotlib
import warnings
import sys
from matplotlib import cm
import seaborn as sns

plt.rcParams['figure.dpi'] = 200 
```

```python
parameters = {'y_offset' : 0.4,
              'x_offset' : 0.6,
              'canvas_dimension' : {'figsize':(3.5,3.5)},
              'x_title_dimension': {'size':12},
              'y_title_dimension': {'size':12},
              'z_title_dimension': {'size':12}}

labels = {'fontproperties' : fm.FontProperties(size=8),
          'rotation'       : 'vertical'}

ticks_params =          {
                        'direction'      : 'in',
                        'width'          : 1,
                        'color'          : '#A9A9A9',
                        'labelsize'      : 8
                        }

minor_ticks_params =    {
                        'which'      : 'minor',
                        'width'      :  1,
                        'color'      : '#A9A9A9',
                        'direction'  : 'in',
                        }
```

```python
def create_csv_dict(csv_table):

    with open(csv_table, 'r') as ct:
        dataframe = pd.read_csv(ct, header=0)

    return dataframe
```

```python
print(os.getcwd())
csv_file = '/home/leonardo/Desktop/JupyterLab_Templates/csv/kr2_shifts.csv'
csv_dict = create_csv_dict(csv_file)
```

```python
models = list(csv_dict.iloc[:,0])
```

```python
ylim = None # (-8.0, 2.0, 2.0)
xlim = (-10.0, 2.0, 2.0)

fig, ax = plt.subplots(**parameters['canvas_dimension'])
ax.tick_params(axis='x', **ticks_params)
ax.tick_params(axis='y', **ticks_params)
ax.xaxis.set_major_formatter(FormatStrFormatter('$%.1f$'))

if ylim != None:
    ax.set_yticks(np.arange(ylim[0], ylim[1]+parameters['y_offset'], ylim[2]))
    ax.set_ylim(ylim[0]-parameters['y_offset'], ylim[1]+parameters['y_offset'])

if xlim != None:
    ax.set_xticks(np.arange(xlim[0], xlim[1]+parameters['x_offset'], xlim[2]))
    ax.set_xlim(xlim[0]-parameters['x_offset'], xlim[1]+parameters['x_offset'])

# Simple condition to color according 
# to blue or red-shift 
values = list(csv_dict.iloc[:,1])
colors = ['#4E88D5' if (x<0) else '#F86C6C' for x in values]

bar_width = 0.50
ax.barh(models, values, bar_width, edgecolor="#000000", lw=0.5, color=colors, align='center')
ax.axvline(x=0, color='#000000', lw=0.5)

ax.set_yticks(models)
ax.set_yticklabels(models)
ax.invert_yaxis() # labels read top-to-bottom (from matplotlib docu.)

plt.savefig('/home/leonardo/Desktop/blue_red_shift.png', bbox_inches='tight', dpi=600, transparent=True)
```
