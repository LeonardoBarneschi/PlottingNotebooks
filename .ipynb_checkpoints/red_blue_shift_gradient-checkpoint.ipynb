```python
def create_csv_dict(csv_table):

    with open(csv_table, 'r') as ct:
        dataframe = pd.read_csv(ct, header=0)

    return dataframe
```

```python
print(os.getcwd())
csv_file = 'kr2_shifts.csv'
csv_dict = create_csv_dict(csv_file)
```
