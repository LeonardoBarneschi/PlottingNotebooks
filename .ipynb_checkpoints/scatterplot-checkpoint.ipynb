```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
gen_params = {'canvas_dimension' : {'figsize':(5,5)},
              'x_title_dimension': {'size':12},
              'y_title_dimension': {'size':12},
              'z_title_dimension': {'size':12},
              'horizontal_line'  : {'color':'black', 'lw':0.5, 'ls':'--', 'zorder':1}}

caspt2_plot= {'S0' : {'color' : '#00008B', 'lw' : 1, 'zorder' : 0, 'alpha' : 0.8},
              'S1' : {'color' : '#008000', 'lw' : 1, 'zorder' : 0, 'alpha' : 0.8},
              'S2' : {'color' : '#A52A2A', 'lw' : 1, 'zorder' : 0, 'alpha' : 0.8}}

caspt2_scat= {'S0' : {'color' : '#00008B', 'marker' : 'o' ,'zorder' : 4, 'alpha' : 1},
              'S1' : {'color' : '#008000', 'marker' : 'o' ,'zorder' : 4, 'alpha' : 1},
              'S2' : {'color' : '#A52A2A', 'marker' : 'o' ,'zorder' : 4, 'alpha' : 1}}
```

```python
path = '/home/leonardo/Desktop'
csv = '/home/leonardo/Desktop/Pipelines/DISPLACEMENT_PIPELINE/g_results_2/results.csv'
df = pd.read_csv(csv)
```

```python
df = df.sort_values('psb-BLA')
```

```python
df
```

```python
fig, ax = plt.subplots()#**gen_params['canvas_dimension'])
```

```python
ax.plot(df['psb-BLA'], df['S0 CASSCF Energy'],**caspt2_plot['S0'], label='_nolegend_')
ax.plot(df['psb-BLA'], df['S1 CASSCF Energy'],**caspt2_plot['S1'], label='_nolegend_')
ax.plot(df['psb-BLA'], df['S2 CASSCF Energy'],**caspt2_plot['S2'], label='_nolegend_')

#ax.scatter(df['psb-BLA'], df['S0 CASPT2 Energy'],**caspt2_scat['S0'], label='S0 CASPT2')
#ax.scatter(df['psb-BLA'], df['S1 CASPT2 Energy'],**caspt2_scat['S1'], label='S1 CASPT2')
#ax.scatter(df['psb-BLA'], df['S2 CASPT2 Energy'],**caspt2_scat['S2'], label='S2 CASPT2')

ax.set_xlabel('PSB-BLA(A)')
ax.set_ylabel('Total Energy (a.u.)')

ax.scatter(df['psb-BLA'], df['S0 CASSCF Energy'],**caspt2_scat['S0'], label='S0 CASSCF')#, label='_nolegend_')
ax.scatter(df['psb-BLA'], df['S1 CASSCF Energy'],**caspt2_scat['S1'], label='S1 CASSCF')#, label='_nolegend_')
ax.scatter(df['psb-BLA'], df['S2 CASSCF Energy'],**caspt2_scat['S2'], label='S2 CASSCF')#, label='_nolegend_')
ax.legend(ncol=3)
fig.savefig(path+'/'+'casscf.png', bbox_inches='tight', dpi=300)
```
